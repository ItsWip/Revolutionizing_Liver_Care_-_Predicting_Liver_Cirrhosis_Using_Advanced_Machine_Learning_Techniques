```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
csv_file= "processed_dataset_liver.csv" 
df = pd.read_csv(csv_file)
display(df.info())
display(df.head())
```

```python
missing_values = df.isnull().sum()
display(missing_values[missing_values > 0])
```

```python
display(df.describe())
```

```python
#target distribution
plt.figure(figsize=(6, 4))
sns.countplot(x=df.iloc[:, -4])  # Assuming target variable is in 39th column
plt.title("Liver Cirrhosis Diagnosis Distribution")
plt.xlabel("Diagnosis (0: No Cirrhosis, 1: Cirrhosis Present)")
plt.ylabel("Number of Patients")
plt.xticks([0, 1], ['No Cirrhosis', 'Cirrhosis'])
plt.show()
```

```python
#feature correlation
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), cmap="coolwarm", annot=False, linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
df.rename(columns={"Place(location where the patient lives)": "Patient Location"}, inplace=True)
df.rename(columns={"Duration of alcohol consumption(years)":"Alcohol Duration"}, inplace=True)
df.rename(columns={"Quantity of alcohol consumption (quarters/day)": "Alcohol Qty (q/day)"}, inplace=True)
df.rename(columns={"Quantity of alcohol consumption (quarters/day)": "Alcohol (q/day)"}, inplace=True)
df.rename(columns={"Type of alcohol consumed": "Alcohol Type"}, inplace=True)
df.rename(columns={"USG Abdomen (diffuse liver or not)": "USG Liver Diffuse"}, inplace=True)
df.rename(columns={"Predicted Value(Out Come-Patient suffering from liver cirrosis or not)": "Predicted Outcome"}, inplace=True)
df.hist(figsize=(15, 12), bins=30, edgecolor='black')
plt.suptitle("Feature Distributions", fontsize=16, y=1.05)  
plt.tight_layout() 
plt.show()
```

```python
categorical_features = ['Gender', 'Hepatitis B infection', 'Hepatitis C infection', 'Diabetes Result', 'Obesity']
for feature in categorical_features:
    plt.figure(figsize=(6, 4))
    sns.countplot(x=df[feature])
    plt.title(f"Distribution of {feature}")
    plt.show()
```

```python
selected_features = ['Age', 'Albumin (g/dl)', 'Total Bilirubin (mg/dl)', 'SGOT/AST (U/L)', 'SGPT/ALT (U/L)']
sns.pairplot(df[selected_features])
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['Albumin (g/dl)'], y=df['Total Bilirubin (mg/dl)'], hue=df.iloc[:, -4])
plt.title("Albumin vs Total Bilirubin Levels")
plt.xlabel("Albumin (g/dl)")
plt.ylabel("Total Bilirubin (mg/dl)")
plt.show()
```
