```python
df['Gender'] = df['Gender'].str.strip().str.lower()
df['Gender'] = df['Gender'].replace({'female ': 'female'})
```

```python
df['Place(location where the patient lives)'] = df['Place(location where the patient lives)'].str.strip().str.lower()
df['Place(location where the patient lives)'] = df['Place(location where the patient lives)'].replace({' ocun': 'unknown'})
```

```python
df['Type of alcohol consumed'] = df['Type of alcohol consumed'].str.strip().str.lower()
df['Type of alcohol consumed'] = df['Type of alcohol consumed'].replace({' branded liquor': 'branded liquor'})
```

```python
df['Hepatitis B infection'] = df['Hepatitis B infection'].str.lower()
df['Hepatitis B infection'] = df['Hepatitis B infection'].replace({'Positive': 'positive'})

df['Hepatitis C infection'] = df['Hepatitis C infection'].str.lower()
df['Hepatitis C infection'] = df['Hepatitis C infection'].replace({'Positive': 'positive'})
```

```python
df['Diabetes Result'] = df['Diabetes Result'].str.upper()
```

```python
df['Obesity'] = df['Obesity'].str.lower()
```

```python
df['Family history of cirrhosis/ hereditary'] = df['Family history of cirrhosis/ hereditary'].str.lower()
df['Family history of cirrhosis/ hereditary'] = df['Family history of cirrhosis/ hereditary'].replace({'husband': 'yes'})
```

```python
df[['Systolic BP', 'Diastolic BP']] = df['Blood pressure (mmhg)'].str.split('/', expand=True)
df['Systolic BP'] = pd.to_numeric(df['Systolic BP'], errors='coerce')
df['Diastolic BP'] = pd.to_numeric(df['Diastolic BP'], errors='coerce')
df.drop(columns=['Blood pressure (mmhg)'], inplace=True)
```

```python
df['Quantity of alcohol consumption (quarters/day)'] = pd.to_numeric(df['Quantity of alcohol consumption (quarters/day)'], errors='coerce')
df.loc[df['Quantity of alcohol consumption (quarters/day)'] > 10, 'Quantity of alcohol consumption (quarters/day)'] = None
```

```python
df['TG'] = pd.to_numeric(df['TG'], errors='coerce')
```

```python
df['LDL'] = pd.to_numeric(df['LDL'], errors='coerce')
```

```python
df.drop(columns=['S.NO'], inplace=True)
```

```python
numeric_columns = ['TCH', 'TG', 'LDL', 'HDL', 'Hemoglobin  (g/dl)', 'PCV  (%)', 
                   'RBC  (million cells/microliter)', 'MCV   (femtoliters/cell)',
                   'MCH  (picograms/cell)', 'MCHC  (grams/deciliter)', 'Total Count',
                   'Polymorphs  (%) ', 'Lymphocytes  (%)']

df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')
```

```python
num_cols = df.select_dtypes(include='number').columns
df[num_cols]= df[num_cols].fillna(df[num_cols].median())
```

```python
cat_cols = df.select_dtypes(include='object').columns
df[cat_cols]= df[cat_cols].fillna(df[cat_cols].mode().iloc[0])
```

```python
df[num_cols] = df[num_cols].applymap(lambda x: str(x).replace('o', '0') if isinstance(x, str) else x)
```

```python
df.columns = df.columns.str.strip().str.replace(r'\s+', ' ', regex=True)
```

```python
categorical_cols = ['Gender', 'Place(location where the patient lives)', 'Type of alcohol consumed', 
                    'Hepatitis B infection', 'Hepatitis C infection', 'Diabetes Result', 
                    'Obesity', 'Family history of cirrhosis/ hereditary', 'USG Abdomen (diffuse liver or not)', 
                    'Predicted Value(Out Come-Patient suffering from liver cirrosis or not)']

encoder = LabelEncoder()
for col in categorical_cols:
    df[col] = encoder.fit_transform(df[col])
```

```python
df[categorical_cols].head()
```

```python
import re

def clean_value(value):
    if isinstance(value, str):
        value = value.strip() 
        value = re.sub(r'\.$', '', value)  
        if re.match(r'^\d+:\d+:\d+$', value):  
            return None  
    return value

df["A/G Ratio"] = df["A/G Ratio"].apply(clean_value)
```

```python
df["A/G Ratio"] = pd.to_numeric(df["A/G Ratio"], errors='coerce')
```

```python
print(df["A/G Ratio"].isnull().sum())
df["A/G Ratio"].fillna(df["A/G Ratio"].mean(), inplace=True)
```

```python
df.rename(columns={'Hemoglobin  (g/dl)': 'Hemoglobin (g/dl)'}, inplace=True)
```

```python
df.replace({'o': '0'}, regex=True, inplace=True)
```

```python
num_cols = ['Age', 'Duration of alcohol consumption(years)', 'Quantity of alcohol consumption (quarters/day)',
            'TCH', 'TG', 'LDL', 'HDL', 'Hemoglobin (g/dl)', 'PCV (%)', 'RBC (million cells/microliter)', 
            'MCV (femtoliters/cell)', 'MCH (picograms/cell)', 'MCHC (grams/deciliter)', 'Total Count', 
            'Platelet Count (lakhs/mm)', 'Total Bilirubin (mg/dl)', 'Direct (mg/dl)', 'Indirect (mg/dl)', 
            'Total Protein (g/dl)', 'Albumin (g/dl)', 'Globulin (g/dl)', 'A/G Ratio', 'AL.Phosphatase (U/L)', 
            'SGOT/AST (U/L)', 'SGPT/ALT (U/L)']

scaler = StandardScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])
```

```python
for col in num_cols:
    non_numeric = df[col][df[col].apply(lambda x: not str(x).replace('.', '', 1).isdigit())] 
    if not non_numeric.empty:
        print(f"Column '{col}' has non-numeric values: {non_numeric.unique()}")
```

```python
df[num_cols] = df[num_cols].apply(pd.to_numeric, errors='coerce')
```

```python
df.fillna(df.mean(), inplace=True)
```

```python
print(df['Predicted Value(Out Come-Patient suffering from liver cirrosis or not)'].value_counts())
```

```python
df_majority = df[df['Predicted Value(Out Come-Patient suffering from liver cirrosis or not)'] == 0]
df_minority = df[df['Predicted Value(Out Come-Patient suffering from liver cirrosis or not)'] == 1]

df_minority_upsampled = resample(df_minority, 
                                 replace=True,
                                 n_samples=len(df_majority),
                                 random_state=42) 
df_upsampled = pd.concat([df_majority, df_minority_upsampled])
df_upsampled = df_upsampled.sample(frac=1, random_state=42).reset_index(drop=True)
print(df_upsampled['Predicted Value(Out Come-Patient suffering from liver cirrosis or not)'].value_counts())
```
